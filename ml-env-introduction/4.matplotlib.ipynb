```python
# 准备工作，读取数据
import pandas as pd
df = pd.read_csv('../data/gapminder.csv')
gdp_cap = df.gdp_cap.tolist()
life_exp = df.life_exp.tolist()
pop = df.population.tolist()
```

查看数据

```python
print(gdp_cap)
print(life_exp)
print(pop)
```

散点图：GDP和寿命的关系

```python
import matplotlib.pyplot as plt

plt.plot(gdp_cap, life_exp)
plt.show()
```

```python
plt.scatter(gdp_cap, life_exp)
plt.xscale('log')
plt.show()
```
